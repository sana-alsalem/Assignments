```python
from datetime import datetime
#convert date_ column to numeric values
dates = dataframe.index
dates_tonumeric = dates.map(lambda date: date.toordinal())

#adding the two next months
last_date = dates_tonumeric[-1]
next_2months = []
for i in range(1, 62):
  next_2months.append(last_date + i)
dataframe['numeric_date'] = dates_tonumeric
dataframe.tail()
```

```python


# Create the DataFrame
predicted_watches = pd.DataFrame(columns=['dates', 'Total_watch_time_in_houres'])
predicted_watches['numeric_date'] = next_2months
#creating linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x = dataframe['numeric_date'].values.reshape(-1, 1)
y = dataframe['Total_watch_time_in_houres']

#spilting the date
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
#fitting the model
model.fit(x_train, y_train)

#predicting the model
predicted_watches['Total_watch_time_in_houres'] = model.predict(predicted_watches['numeric_date'].values.reshape(-1, 1))

```

```python
#convert the ordinal date to the original date
predicted_watches['date'] = predicted_watches['numeric_date'].apply(lambda x: datetime.fromordinal(x))
predicted_watches.set_index('date', inplace=True)
predicted_watches.drop('dates', axis=1, inplace=True)
predicted_watches.head()
```
